```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/cafe_vendas.csv')
df.head()
```

```python
df.columns = ['ID Transacao', 'Item', 'Quantidade', 'Preco Unitario', 'Gasto Total',
              'Metodo Pagamento', 'Localidade', 'Data Transacao']
```

```python
df.head()
```

Traduzindo a coluna `Item` para português

```python
df['Item'].unique()
```

```python
item_br = ['Cafe', 'Bolo', 'Cookie', 'Salada', 'Smoothie',
           'Nao Sabemos', 'Sanduiche', np.nan, 'ERRO', 'Suco', 'Cha']

x = df['Item'].unique()

for i in range(len(x)):

    df.loc[df['Item'] == x[i], 'Item'] = item_br[i]
```

```python
df['Item'].unique()
```

Traduzindo a coluna `Metodo Pagamento` para português

```python
df['Metodo Pagamento'].unique()
```

```python
pagamento_br = ['Cartao de Credito', 'Dinheiro', 'Nao Sabemos',
                'Carteira Digital', 'ERRO', np.nan]

y = df['Metodo Pagamento'].unique()

for i in range(len(y)):

    df.loc[df['Metodo Pagamento'] == y[i], 'Metodo Pagamento'] = pagamento_br[i]

print(df['Metodo Pagamento'].unique())
```

Traduzindo a coluna `Localidade` para português

```python
df['Localidade'].unique()
```

```python
local_br = ['Entrega', 'Loja', 'Nao Sabemos', np.nan, 'ERRO']

z = df['Localidade'].unique()

for i in range(len(z)):

    df.loc[df['Localidade'] == z[i], 'Localidade'] = local_br[i]

print(df['Localidade'].unique())
```

Quero substituir todos os ERROR nas colunas que ainda não traduzi por ERRO e traduzir os UNKNOWN por Nao Sabemos

```python
colunas = ['ID Transacao','Quantidade', 'Preco Unitario', 'Gasto Total', 'Data Transacao']

for col in colunas:

    df.loc[df[col] == 'ERROR', col] = 'ERRO'

    df.loc[df[col] == 'UNKNOWN', col] = 'Nao Sabemos'
```

```python
df.head(10)
```

```python
df.to_csv('vendas_cafe.csv', index = False)
```
